```python
import matplotlib.pyplot as plt
import numpy as np

# Function: Linear Congruential Generator (LCG)
def prng_lcg(seed=42, a=1664525, c=1013904223, m=2**32, n=10000):
    x = seed
    result = []
    for _ in range(n):
        x = (a * x + c) % m
        result.append(x / m)  # Normalize to [0, 1)
    return result

# Generate PRNG values
prng_values = prng_lcg()

# --- Plot 1: Uniformity Test ---
plt.figure(figsize=(10, 4))
plt.hist(prng_values, bins=50, color='skyblue', edgecolor='black')
plt.title("Uniformity Test - Histogram of PRNG Values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# --- Plot 2: Dependency Test ---
plt.figure(figsize=(6, 6))
plt.scatter(prng_values[:-1], prng_values[1:], s=1, alpha=0.5)
plt.title("Dependency Test - Scatter Plot (X[n] vs X[n+1])")
plt.xlabel("X[n]")
plt.ylabel("X[n+1]")
plt.grid(True)
plt.show()
```
